Data fields

• Total Spend in Months 1 and 2 of 2017: The total spend of a customer in the months July & August 2017.

• Total SMS Spend: The total spend on SMS by a customer revenue earned through the SMS service used by the subscriber.

• Total Data Spend: The total amount spend on Data/Internet by a customer revenue earned through the SMS service used by the subscriber.

• Total Data Consumption : The total data consumed by a subscriber in KB over the period under study

• Total Unique Calls: The total count of unique calls made by a subscriber during the period under review

• Total Onnet spend: The total spend of a customer to make on-network calls (on the same network as the subscriber)

• Total Offnet spend: Aggregate of Off Net Revenue: The total spend of a customer to make off-network calls (not the same network as the subscriber)

• Customer Tenure in Months: The time passed since the subscriber started using the services of the network provider and counted in months

• Network type subscription in Month 1: This indicates preferred network subscription of a customer, which may indicate their type of device - 2G or 3G service.

• Network type subscription in Month 2: This indicates preferred network subscription of a customer, which may indicate their type of device - 2G or 3G service in the month after.

• Total Call centre complaint calls: The number of complaints made by the subscribers.

• Most Loved Competitor network in in Month 1: This information certainly have a huge impact on as it gives the information about which other service provider prefer and may likely move to

• Most Loved Competitor network in in Month 2: This information certainly have a huge impact on as it gives the information about which other service provider prefer and may likely move to

• Churn Status: This is an indicator of a customer who leaves the network or not. 1 means the customer has churned and 0 means no churn.

<p style="font-family: Arial; font-size:1.75em;color:purple; font-style:bold"><br>

Features Engineering, Selection and Targets Extraction
<br><br></p>


```python
# calculate as a 30 day period
train['Customer tenure in month'] = np.floor(train['Customer tenure in month']*30).astype(int)
```

```python
train['Data Spend Consumption'] = pd.qcut((train['Total Data Spend'] * train['Total Data Consumption']), 3, labels=False)
```

```python
train['Total Data Consumption'] = pd.qcut(train['Total Data Consumption'], 6, labels=False)
```

```python
train['Total Spend in Months 1 and 2 of 2017'] = np.log10(train['Total Spend in Months 1 and 2 of 2017']).round(decimals=2)
```

```python
train['Total Onnet spend'] = pd.qcut(train['Total Onnet spend'],3, labels=False)
train['Total Offnet spend'] = pd.qcut(train['Total Offnet spend'],3, labels=False)
```

```python
train['Spend on Calls'] = train['Total Onnet spend'] * train['Total Offnet spend']
```

```python
train = pd.get_dummies(train, columns=['Most Loved Competitor network in in Month 2'])
```

```python
pp.figure(figsize=(11,6))
sb.heatmap(train.corr(),cmap='rainbow_r',annot=True, annot_kws = {'fontsize' : 10})  # heatmap feature correlation matrix
pp.show()
```

```python
cols = train.drop('Churn Status',axis='columns').columns
cols
```

```python
correlations = [train['Churn Status'].corr(train[f]) for f in cols]
correlations
```

```python
# create a function for plotting a dataframe with string columns and numeric values
def plot_dataframe(frame, label):  
    color='coral'
    fig = pp.gcf()
    fig.set_size_inches(10,5)
    pp.ylabel(label)

    ax = frame.correlation.plot(linewidth=3, color=color)
    ax.set_xticks(frame.index)
    ax.set_xticklabels(frame.attributes, rotation=90)
    pp.show()
    
#create a dataframe using cols and correlations
frame = pd.DataFrame({'attributes': cols, 'correlation': correlations})

# let's plot above dataframe using the function we created  
plot_dataframe(frame, 'Correlated to Performance')
```

Choose the Engineered Features to Throw into the Model and Extract the Targets Values

```python
X = features_data                # Input
print(X.shape)
X[:3]
```

<p style="font-family: Arial; font-size:1.75em;color:purple; font-style:bold"><br>

Split the Dataset into Training and Test Datasets
<br><br></p>

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=132)
```

<p style="font-family: Arial; font-size:1.75em;color:purple; font-style:bold"><br>

Building a Model To Predict the Repayment/Default of these customers
<br><br></p>


<p style="font-family: Arial; font-size:1.75em;color:purple; font-style:bold"><br>

(1) Logistic Regression
<br><br></p>


```python
from sklearn.linear_model import LogisticRegression         # 1. Choose the model class
lr = LogisticRegression()                                   # 2. Instantiate the model
lr.fit(X_train, y_train)                                    # 3. Fit the data to the model
```

```python
# ?LogisticRegression()
```

```python
lr_prediction = lr.predict(X_test)                                   # 4. Predict targets of Test data
```

```python
lr.coef_,lr.intercept_
```

```python
from sklearn.metrics import accuracy_score, r2_score
accuracy = accuracy_score(y_test,lr_prediction)                      # 5. Evaluate the model's performance             
print('Accuracy of Logistic Regression is:',accuracy) 
```

```python
r2_score(y_test,lr_prediction)
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test,lr_prediction)                 # 5. Evaluate the model's performance             
print(report) 
```

```python
from sklearn.model_selection import cross_val_score                  # 5. Validate the model's performance
cross_val_score(lr, X_train, y_train, cv=5)
```

```python
from sklearn.metrics import confusion_matrix
confusion = confusion_matrix(y_test,lr_prediction)                   # 5. Evaluate the model's performance             
sb.heatmap(confusion ,square=True, annot=True, cbar=False)
pp.xlabel('true value')
pp.ylabel('predicted value');
```
